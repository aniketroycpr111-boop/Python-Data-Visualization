## Closing Price Over Time (Line Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = 'Google_stock.csv'
df =pd.read_csv(file_path)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Sort values
df = df.sort_values('Date')

# Plot
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], color='blue')
plt.title("Google Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```

## Trading Volume Over Time (Bar Chart)

```python
plt.figure(figsize=(12,6))
plt.bar(df['Date'], df['Volume'], color='orange')
plt.title("Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

## Daily Return Percentage Distribution (Histogram)

```python
plt.figure(figsize=(8,5))
plt.hist(df['Daily_Return_Pct'], bins=50, color='green')
plt.title("Distribution of Daily Returns (%)")
plt.xlabel("Daily Return (%)")
plt.ylabel("Frequency")
plt.show()
```

## Moving Average Trend (Very Important

```python
# Create Moving Averages
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_200'] = df['Close'].rolling(window=200).mean()

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], label='Close Price', alpha=0.5)
plt.plot(df['Date'], df['MA_50'], label='50-Day MA', color='red')
plt.plot(df['Date'], df['MA_200'], label='200-Day MA', color='green')

plt.title("Close Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

## Monthly Average Closing Price (Trend by Month)

```python
monthly_avg = df.groupby('Month')['Close'].mean()

plt.figure(figsize=(8,5))
monthly_avg.plot(kind='bar', color='red')

plt.title("Average Closing Price by Month")
plt.xlabel("Month")
plt.ylabel("Average Close Price")
plt.xticks(rotation=0)
plt.show()
```

## Closing Price Over Time (Seaborn Lineplot)

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x='Date', y='Close', data=df, color='blue')

plt.title("Google Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

## Daily Return Distribution (Histogram + KDE)

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Daily_Return_Pct'], bins=50, kde=True, color='green')

plt.title("Distribution of Daily Returns (%)")
plt.xlabel("Daily Return (%)")
plt.ylabel("Frequency")
plt.show()
```

## Boxplot of Returns by Day of Week

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Day_of_Week', y='Daily_Return_Pct', data=df)

plt.title("Daily Return by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Daily Return (%)")
plt.xticks(rotation=45)
plt.show()
```
